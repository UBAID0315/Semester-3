```python
import webbrowser
import speech_recognition as sr
import os
import pyttsx3
import subprocess
import pyautogui
import random
import platform
import time
import psutil
import wikipediaapi

wiki_wiki = wikipediaapi.Wikipedia('en',extract_format=wikipediaapi.ExtractFormat.WIKI, headers={'User-Agent': 'Your-User-Agent-Here'})

def search_wikipedia(search_query):
        page_py = wiki_wiki.page(search_query)
        if page_py.exists():
            wikipedia_url = page_py.fullurl
            webbrowser.get('chrome').open_new(wikipedia_url)
            definition = page_py.text[:300]
            speak(f"Here is a brief definition of {search_query}: {definition}")
        else:
            speak(f"Sorry, I couldn't find information about {search_query} on Wikipedia.")

def sing_me_a_song():
    songs = [
        "Twinkle, twinkle, little star, How I wonder what you are.",
        "Happy birthday to you, happy birthday to you!",
        "I'm a little teapot, short and stout, here is my handle, here is my spout.",
        # Add more song lyrics or customize as needed
    ]

    chosen_song = random.choice(songs)
    speak("Sure, here's a song for you.")
    speak(chosen_song, rate=140)  # Adjust rate for singing-like effect

# Speech engine initialization
engine = pyttsx3.init()
voices = engine.getProperty('voices')
engine.setProperty('voice', voices[0].id)  # 0 = male, 1 = female
activationWord = 'computer'  # Single word

# Set the browser
chrome_path = r"C:\Program Files\Google\Chrome\Application\chrome.exe"
webbrowser.register('chrome', None, webbrowser.BackgroundBrowser(chrome_path))

def speak(text, rate=170):
    engine.setProperty('rate', rate)
    engine.say(text)
    engine.runAndWait()


def listOrDict(var):
    if isinstance(var, list):
        return var[0]['plaintext']
    else:
        return var['plaintext']
def get_system_information():
    system_info = {
        "Operating System": platform.system(),
        "OS Version": platform.version(),
        "Processor": platform.processor(),
        "Architecture": platform.architecture(),
        "RAM": f"{psutil.virtual_memory().total / (1024 ** 3):.2f} GB",
        "Storage": f"{psutil.disk_usage('/').total / (1024 ** 3):.2f} GB",
    }

    info_text = "Here is information about your system:\n"
    for key, value in system_info.items():
        info_text += f"{key}: {value}\n"

    speak(info_text)

def listen_for_command():
    recognizer = sr.Recognizer()

    with sr.Microphone() as source:
        print("Listening for a command...")
        speak("I am listening to you")
        recognizer.adjust_for_ambient_noise(source)
        audio = recognizer.listen(source, timeout=8)

    try:
        print("Recognizing speech...")
        command = recognizer.recognize_google(audio)
        print(f"Command: {command}")
        return command.lower()  # Convert the command to lowercase for easier comparison
    except sr.UnknownValueError:
        print("Could not understand the audio")
    except sr.RequestError as e:
        print(f"Could not request results from Google Web Speech API; {e}")

def perform_action(command):

    if "open" in command and "notepad" in command:
        print("Opening Notepad...")
        os.system("notepad.exe")
        speak("Opening Notepad")
        time.sleep(5)
    
    elif "open" in command and "chrome" in command:
        print("Opening Google Chrome...")
        path = r"C:\Program Files\Google\Chrome\Application\chrome.exe"
        os.system(f'"{path}"')
        speak("Opening Google Chrome")
        time.sleep(5)
    
    elif "open" in command and "google" in command:
        print("Opening Google for you")
        webbrowser.get('chrome').open_new('https://www.google.com.pk')
        speak("Opening Google")
        time.sleep(5)
    
    elif "open" in command and "steam" in command:
        print("Opening Steam for you")
        speak("Opening Steam for you")
        steam_path = r"C:\Program Files (x86)\Steam\steam.exe" 
        os.system(f'"{steam_path}"')
        time.sleep(5)

    elif "open" in command and "calculator" in command:
        print("Opening Calculator...")
        calc_app_path = r"C:\\Windows\\System32\\calc.exe" 
        subprocess.Popen(calc_app_path)
        time.sleep(5)

    elif "open" in command and "clock" in command:
        print("Opening Clock...")
        os.system("start ms-clock:")
        speak("Opening Clock")
        time.sleep(5)

    elif "sing me a song" in command:
        sing_me_a_song()
        time.sleep(3)

    elif "tell me about my system" in command:
        get_system_information()
        time.sleep(5)

    elif "search wikipedia for" in command:
        search_query = command.replace("search wikipedia for", "").strip()
        search_wikipedia(search_query)
        time.sleep(5)
    
    elif "open youtube for" in command:
        search_query = command.replace("open youtube for", "").strip()
        print(f"Searching YouTube for: {search_query}")
        
        youtube_url = f"https://www.youtube.com/results?search_query={search_query}"

        webbrowser.get('chrome').open_new(youtube_url)
        speak(f"Opening YouTube and searching for {search_query}.")
        time.sleep(5)

    elif "close notepad" in command:
        print("Closing the opened app")
        os.system("taskkill /f /im notepad.exe")
        speak("Closing")

    elif "close calculator" in command:
        print("Closing the opened app")
        pyautogui.hotkey("alt","f4")
        speak("Closing")

    elif "close chrome" in command:
        print("Closing the opened app")
        os.system("taskkill /f /im chrome.exe")
        speak("Closing")

    elif "close steam" in command:
        print("Closing the opened app")
        os.system("taskkill /f /im steam.exe")
        speak("Closing")

    elif "close clock" in command:
        print("Closing the opened app")
        os.system("taskkill /f /im ms-clock.exe")
        speak("Closing")

    elif "close google" in command:
        print("Closing the opened app")
        pyautogui.hotkey("ctrl","w")
        speak("Closing")
    
    elif "exit" in command or "close" in command or "goodbye" in command:
        print("Exiting program...")
        speak("Okay!Have a good day")
        return True

    else:
        print("Command not recognized.")
    return False

# Main Function
if __name__ == "__main__":
    exit_program = False

    while not exit_program:
        command = listen_for_command()
        if command:
                exit_program = perform_action(command)
```
